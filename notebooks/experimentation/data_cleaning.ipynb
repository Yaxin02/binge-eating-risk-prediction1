```python
import numpy as np

target_cols = [
    'Perception_EatingDisorder_Possibly, but I am not sure',
    'Perception_EatingDisorder_Probably not',
    'Perception_EatingDisorder_Probably yes',
    'Perception_EatingDisorder_Yes, I believe I do'
]

df_encoded["target"] = df_encoded[target_cols].idxmax(axis=1)
```

```python
df_encoded["target"] = df_encoded["target"].str.replace("Perception_EatingDisorder_", "")
```

```python
df_encoded = df_encoded.drop(columns=target_cols)
```

```python
print("Shape:", df_encoded.shape)
print("\nFirst rows:")
display(df_encoded.head())
```

```python
print("==== FINAL DATA CHECK ====")

print("\nShape:", df_encoded.shape)

print("\nMissing values:", df_encoded.isnull().sum().sum())

print("\nDuplicates:", df_encoded.duplicated().sum())

print("\nObject columns:", df_encoded.select_dtypes(include=["object"]).columns.tolist())

print("\nTarget distribution:")
print(df_encoded["target"].value_counts())
```
